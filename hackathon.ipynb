```python
# fill missing values with scikit-learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer


# Fill categorical values with 'missing' & numerical values with mean
cat_imputer=SimpleImputer(strategy="constant", fill_value="missing")
# door_imputer =SimpleImputer(strategy="constant", fill_value=4)
# num_imputer=SimpleImputer(strategy="mean")

xyz_concern_imputer=SimpleImputer(strategy="constant", fill_value=int(2))
xyz_knowledge_imputer=SimpleImputer(strategy="constant", fill_value=int(1))
behavioral_face_mask_imputer=SimpleImputer(strategy="constant",fill_value=int(0))
opinion_xyz_vacc_effective_imputer=SimpleImputer(strategy="constant", fill_value=int(4))
opinion_seas_risk_imputer=SimpleImputer(strategy="constant", fill_value=int(3))
education_imputer=SimpleImputer(strategy="constant", fill_value="College Graduate")
income_poverty_imputer=SimpleImputer(strategy="constant", fill_value="<= $75,000, Above Poverty")
marital_status_imputer=SimpleImputer(strategy="constant", fill_value="Married")
rent_or_own_imputer=SimpleImputer(strategy="constant", fill_value="Own")
employment_status_imputer=SimpleImputer(strategy="constant", fill_value="Employed")

# Define columns

# door_feature=["Doors"]
# num_features=["Odometer (KM)"]
rent_or_own_features=["rent_or_own"]

xyz_concern_features=["xyz_concern", "opinion_xyz_risk", "opinion_xyz_sick_from_vacc", "opinion_seas_sick_from_vacc"]
xyz_knowledge_features=["xyz_knowledge","behavioral_avoidance","behavioral_antiviral_meds", "behavioral_wash_hands", "behavioral_touch_face", "household_adults", "household_children"]
behavioral_face_mask_features=["behavioral_face_mask", "behavioral_large_gatherings","behavioral_outside_home", "doctor_recc_xyz", "doctor_recc_seasonal", "chronic_med_condition", "child_under_6_months","health_worker", "health_insurance"]
opinion_xyz_vacc_effective_features=["opinion_xyz_vacc_effective", "opinion_seas_vacc_effective"]
opinion_seas_risk_features=["opinion_seas_risk"]
education_features=["education"]
income_poverty_features=["income_poverty"]
marital_status_features=["marital_status"]
# rent_or_own_features=["rent_or_own"]
employment_status=["employment_status"]
cat_features=["employment_industry","employment_occupation"]

# create an imputer (something that fills missing data)
imputer =ColumnTransformer([
    
#     ("door_imputer", door_imputer, door_feature),
#     ("num_imputer", num_imputer, num_features),
    ("xyz_concern_imputer",xyz_concern_imputer,xyz_concern_features),
    ("xyz_knowledge_imputer", xyz_knowledge_imputer, xyz_knowledge_features),
    ("behavioral_face_mask_imputer", behavioral_face_mask_imputer,behavioral_face_mask_features),
    ("opinion_xyz_vacc_effective_imputer", opinion_xyz_vacc_effective_imputer, opinion_xyz_vacc_effective_features),
    ("opinion_seas_risk_imputer", opinion_seas_risk_imputer, opinion_seas_risk_features),
    ("education_imputer", education_imputer, education_features), 
    ("income_poverty_imputer", income_poverty_imputer, income_poverty_features), 
    ("marital_status_imputer", marital_status_imputer, marital_status_features), 
    ("rent_or_own_imputer", rent_or_own_imputer, rent_or_own_features),
    ("employment_status_imputer", employment_status_imputer, employment_status),
    ("cat_imputer", cat_imputer, cat_features)
   ],
remainder="passthrough")


# Transform the data
filled_x=imputer.fit_transform(x)
filled_x
```
